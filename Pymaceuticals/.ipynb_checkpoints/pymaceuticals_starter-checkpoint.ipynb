```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_mouse_data = pd.merge(study_results, mouse_metadata, on="Mouse ID", how="left")

# Display the data table for preview
combined_mouse_data
```

```python
# Checking the number of mice.
no_mice = len(combined_mouse_data["Mouse ID"].unique())
no_mice
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
combined_mouse_df = combined_mouse_data.sort_values(["Mouse ID", "Timepoint"])

# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. Compare row to previous row.
duplicates = combined_mouse_df.loc[(combined_mouse_df["Mouse ID"] == combined_mouse_df["Mouse ID"].shift(1)) & 
                                    (combined_mouse_df["Timepoint"] == combined_mouse_df["Timepoint"].shift(1))]

duplicates
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID. Return if one of them is true. If both false do not return matching items.
cleaned_mouse_df = combined_mouse_data.loc[(combined_mouse_df["Mouse ID"] != "g989")]
cleaned_mouse_df
```

```python
# Checking the number of mice in the clean DataFrame. Removed 13 rows related to mouse g989. Double check # of unique mice left.
no_mice2 = len(cleaned_mouse_df["Mouse ID"].unique())
no_mice2
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
mean_tumor_volume = cleaned_mouse_df.groupby(["Drug Regimen"], observed=False)["Tumor Volume (mm3)"].mean()
median_tumor_volume = cleaned_mouse_df.groupby(["Drug Regimen"], observed=False)["Tumor Volume (mm3)"].median()
tumor_vol_var = cleaned_mouse_df.groupby(["Drug Regimen"], observed=False)["Tumor Volume (mm3)"].var()
tumor_vol_std_dev = cleaned_mouse_df.groupby(["Drug Regimen"], observed=False)["Tumor Volume (mm3)"].std()
tumor_vol_std_err = cleaned_mouse_df.groupby(["Drug Regimen"], observed=False)["Tumor Volume (mm3)"].sem()

# Assemble the resulting series into a single summary DataFrame.
statistics_summary = pd.DataFrame({
    "Mean Tumor Volume": mean_tumor_volume,
    "Median Tumor Volume": median_tumor_volume,
    "Tumor Volume Variance": tumor_vol_var,
    "Tumor Volume Std. Dev": tumor_vol_std_dev,
    "Tumor Volume Std. Err.": tumor_vol_std_err})
# Display results
statistics_summary
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse. Use double brackets to create instant df
last_timepoint_df = cleaned_mouse_df.groupby(["Mouse ID"])[["Timepoint"]].max()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
max_tumor_grouped_df = pd.merge(last_timepoint_df, cleaned_mouse_df, on=["Mouse ID", "Timepoint"], how='left')
max_tumor_grouped_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
volume_list = []

for treatment in treatments:
    # Locate the rows which contain mice on each drug and get the tumor volumes, returns a series
    tumor_vol_dataf = max_tumor_grouped_df.loc[max_tumor_grouped_df["Drug Regimen"] == treatment, "Tumor Volume (mm3)"]

    #Append to volume list & add subsets
    volume_list.append(tumor_vol_dataf)  
    
    # Calculate the IQR and quantitatively determine if there are any potential outliers.
    # If the data is in a dataframe, we can use pandas to give quartile calculations but must sort data
    
    #volume_list.sort()
    sorted_df = tumor_vol_dataf.sort_values()

    quartiles = sorted_df.quantile([.25,.5,.75])
    lowerq = quartiles[.25]
    upperq = quartiles[.75]
    iqr = upperq-lowerq
    
    # Determine outliers using upper and lower bounds
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)

    outlier_tumor =  tumor_vol_dataf.loc[(tumor_vol_dataf < lower_bound) | (tumor_vol_dataf > upper_bound)]
    print(f"{treatment}'s potential outliers: {outlier_tumor}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group

fig1, ax1 = plt.subplots()
# chatgpt helped me change the marker to red
ax1.boxplot(volume_list, labels=treatments, flierprops=dict(markerfacecolor='r',markersize=10))
ax1.set_ylabel('Final Tumor Volume (mm3)')


plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
single_mouse_df = combined_mouse_data.loc[(combined_mouse_df["Mouse ID"] == "r944")]
# already sorted

plt.plot(single_mouse_df['Timepoint'], single_mouse_df['Tumor Volume (mm3)'], linestyle="-", linewidth=2)
plt.title("Capomulin treatment of mouse r944")
plt.xlabel("Timepoint (days)")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
capomulin_df = combined_mouse_data.loc[(combined_mouse_df["Drug Regimen"] == "Capomulin")]
max_weight_df = capomulin_df.groupby(["Mouse ID"])["Weight (g)"].max()

mean_cap_vol = capomulin_df.groupby(["Mouse ID"], observed=False)["Tumor Volume (mm3)"].mean()
# Or add to df: mean_cap_vol_df = pd.DataFrame({"Mean Tumor Volume (mm3)": mean_cap_vol}) but [[ ]] returns a dataframe

plt.scatter(max_weight_df, mean_cap_vol)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

print(f"The correlation between mouse weight and the average tumor volume is: {round(st.pearsonr(max_weight_df,mean_cap_vol)[0],2)}")

# Perform a linear regression on mouse weight versus tumor size
slope, intercept, rvalue, pvalue, stderr = st.linregress(max_weight_df, mean_cap_vol)

# Create equation of line to calculate predicted number of petrol-electric cars
tumor_line = slope * max_weight_df + intercept

plt.scatter(max_weight_df, mean_cap_vol)
plt.plot(max_weight_df, tumor_line, "r-")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()

# (slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
# regress_values = x_values * slope + intercept
# line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
# plt.scatter(x_values,y_values)
# plt.plot(x_values,regress_values,"r-")
```
